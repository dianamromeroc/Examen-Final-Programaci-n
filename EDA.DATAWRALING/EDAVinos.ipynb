```python

#se realiza la importación de las librerias para leer los archivos
from google.colab import files 
  
  
uploaded = files.upload()


import pandas as pd 
import io 
  
  
```

```python
# se verifica cada una de las variables en el archivo de vino rojo y blanco
red = pd.read_excel(io.BytesIO(uploaded['winequality-red.xlsx'])) 
red
red.head(4)

```

```python
white = pd.read_excel(io.BytesIO(uploaded['winequality-white.xlsx']))
white
white.head(4)
```

```python
# se reliza el importe de las librerias y se incluye una variable llamada 
#categoria_vinos a cada DataFrame que indiqué si el vino es rojo (red) o blanco (white). Esto servirá para distinguir el tipo de vino de cada fila al unir los Datos.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
red['categoria_vinos']='red'
red.head()
```

```python
white['categoria_vinos']='white'
white.head()
```

```python
# se utiliza pandas para lograr la unión de los datos y se verifica que tenga la categoria asignada para distinguir los datos
vinos_red_white=red.append(white, ignore_index=True)
vinos_red_white
```

```python
# verificamos el tipo de dato y si existen nulos
vinos_red_white.info()
```

```python
# utilizamos describe para Cálcular las estadísticas descriptivas del dataset
vinos_red_white.describe()
```

```python
quality = vinos_red_white['quality']
np.unique(quality, return_counts=True)
values, distribucion = np.unique(quality, return_counts=True)
plt.bar(values, distribucion)

# se visualiza con la variable quality que es una variable que más se repite en 6 y
# que promedia la información dado que aproximadamente es una distribución binomial es decir se mantiene en cuanto a calidad del producto
```

```python
# se desea observar los datos estadisticos de quality para realizar una correlación entre esta variable y el nivel de alchol dado si se ve afectado

print("Valor de la moda: " + str(quality.mode()))
print("Valor de la media: " + str(quality.mean()))
print("Valor de la mediana: " + str(quality.median()))
print("Valor Maximo: " + str(quality.max()))
print("Valor Minimo: " + str(quality.min()))


plt.scatter(vinos_red_white["quality"], vinos_red_white["alcohol"])
plt.xlabel("alcohol")
plt.xlabel("quality")
plt.show()
```

```python
# En el anterior grafico se puede observar en la grafica lo correspondiente a la estadistica, en su nivel
# maximo de calidad se ve involucrado el mas alto nivel de alcohol
# de igual manera se verica al contrario de variable y se visualiza de igual manera
alcohol = vinos_red_white['alcohol']
print("Valor de la moda: " + str(alcohol.mode()))
print("Valor de la media: " + str(alcohol.mean()))
print("Valor de la mediana: " + str(alcohol.median()))
print("Valor Maximo: " + str(alcohol.max()))
print("Valor Minimo: " + str(alcohol.min()))

plt.scatter(vinos_red_white["alcohol"], vinos_red_white["quality"])
plt.xlabel("quality")
plt.xlabel("alcohol")
plt.show()
```
